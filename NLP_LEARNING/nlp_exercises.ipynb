```python
import spacy

nlp = spacy.load('en_core_web_sm')
text = "Hi, I am looking for a job! Can you please provide me details"

res = nlp(text)
for sentences in res.sents:
    print(sentences)
```

```python
for sentences in res:
    print(sentences)
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
from nltk.text import sent_tokenize

docs = sent_tokenize(text)
print(docs)
```

```python
from nltk.tokenize import word_tokenize

docs = word_tokenize(text)

print(docs)
```
