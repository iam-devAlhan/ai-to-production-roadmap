> Understanding the concept of Language Processing Pipeline.

SpaCy provide us pretrained models such as en_core_web_sm for english language that allows us to create a pipeline and add components such as POS (Parts of Speech Tagging), Lemmatization, NER (Named Entity Recognition) etc

```python
# Using the preloaded pipeline and outputting characterisitics
nlp = spacy.load("en_core_web_sm")
text = nlp("Imran Khan was the leading Captain of ICC World Cup 1992 and He led Pakistan to a victory against England")

for doc in text:
    print(doc, " | ", doc.pos_, " | ", doc.lemma_)
```

```python
# Finding NER entities in text with explaination
for token in text.ents:
    print(token.text, " | ", token.label_, " | ", spacy.explain(token.label_))
```

```python
from spacy import displacy

displacy.render(text, style="ent")
```

## Stemming and Lemmatization

> **Stemming** is a technique in which words are stemmed into base words, I mean simple words are converted into Base words for example eating to eat, adaptive to adapt.

> Whereas **Lemmatization** is an improved version of stemming which not only looks for suffixes or prefixes but has knowledge of that particular language for example, ate to eat.

Using NLTK we can do both stemming and lemmatization

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["adaptive", "talkative", "processing", "added", "sealed", "forged", "shooting", "drafting"]

for word in words:
    print("Word: " + word, "| Stemmed Word: " +  stemmer.stem(word))
```

# Build a Spam and Ham SMS Classifier using Bag of Words and CountVectorizer

```python
import pandas as pd
df = pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.iloc[0]
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df.rename(columns={"v1": "Class", "v2": "Message"}, inplace=True)
```

```python
df.columns
```

```python
df["Message"] = df["Message"].str.replace(r"[^a-zA-Z0-9_\s]", "", regex=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["Class_Encode"] = le.fit_transform(df["Class"])
```

```python
df.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(df["Message"])

```

```python
X.shape
```

```python
print(cv.get_feature_names_out(X)[::-20])
```

```python
df["Class_Encode"].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df["Class_Encode"])

nb = MultinomialNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
print(classification_report(y_test, y_pred=y_pred))
```

```python
test_messages = [
    "Hey, are we still meeting for lunch at 12? Let me know.",                             # Classic Ham
    "WINNER! You have been selected for a cash prize of $1000. Click here to claim now!",   # Obvious Spam
    "Can you send me the files for the project by tonight?",                               # Work Ham
    "URGENT: Your account has been compromised. Verify your identity at this link.",       # Phishing/Spam
    "Congratulations! You've won a free gift card. Call 0800 now to redeem.",             # Classic Spam
    "I'll be home late, don't wait up for dinner.",
    "I forgot my wallet at lunch, can I borrow $20? I'll pay you back tonight.", 
    
    # 2. THE "URGENT" HAM: Uses high-pressure words but for a real reason.
    "URGENT: Your mom is at the hospital. Call me as soon as you see this.",
    
    # 3. THE "SNEAKY" SPAM: Avoids 'winner' or 'prize', uses friendly language.
    "Hey! It's been a while. I found this link and thought of you: http://bit.ly",
    
    # 4. THE "PROMOTIONAL" HAM: Sounds like an ad, but might be a newsletter the user wants.
    "Your weekly summary of top news is here. Click to read more.",
    
    # 5. THE "SHORT" SPAM: Very few words, hard for CountVectorizer to grab.
    "Claim it now: 0800 123456",
    
    # 6. THE "JOB" SPAM: Common scam that looks like a LinkedIn message.
    "Work from home. Earn $500 per day. No experience needed. Reply YES."
]


test_transformed = cv.transform(test_messages)

# 2. Predict using the trained Naive Bayes model
predictions = nb.predict(test_transformed)

# 3. Print the results nicely
for msg, pred in zip(test_messages, predictions):
    label = "SPAM" if pred == 1 else "HAM"
    print(f"[{label}] -> {msg}")
```
