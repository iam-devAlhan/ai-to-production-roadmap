> Basics of Tokenization

```python
import spacy

nlp = spacy.load('en_core_web_sm')
text = "Hi, I am looking for a job! Can you please provide me details"

res = nlp(text)
for sentences in res.sents:
    print(sentences)
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
from nltk.text import sent_tokenize

docs = sent_tokenize(text)
print(docs)
```

```python
from nltk.tokenize import word_tokenize

docs = word_tokenize(text)

print(docs)
```

> Understanding the concept of Language Processing Pipeline.

SpaCy provide us pretrained models such as en_core_web_sm for english language that allows us to create a pipeline and add components such as POS (Parts of Speech Tagging), Lemmatization, NER (Named Entity Recognition) etc

```python
# Using the preloaded pipeline and outputting characterisitics
nlp = spacy.load("en_core_web_sm")
text = nlp("Imran Khan was the leading Captain of ICC World Cup 1992 and He led Pakistan to a victory against England")

for doc in text:
    print(doc, " | ", doc.pos_, " | ", doc.lemma_)
```

```python
# Finding NER entities in text with explaination
for token in text.ents:
    print(token.text, " | ", token.label_, " | ", spacy.explain(token.label_))
```

```python
from spacy import displacy

displacy.render(text, style="ent")
```

## Stemming and Lemmatization

> **Stemming** is a technique in which words are stemmed into base words, I mean simple words are converted into Base words for example eating to eat, adaptive to adapt.

> Whereas **Lemmatization** is an improved version of stemming which not only looks for suffixes or prefixes but has knowledge of that particular language for example, ate to eat.

Using NLTK we can do both stemming and lemmatization

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["adaptive", "talkative", "processing", "added", "sealed", "forged", "shooting", "drafting"]

for word in words:
    print("Word: " + word, "| Stemmed Word: " +  stemmer.stem(word))
```

```python
docs = nlp("Hey yo, are you in this list. I think i might have messed with ya.!")

for doc in docs:
    print(doc, doc.lemma_)
```
